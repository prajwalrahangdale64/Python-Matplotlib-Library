### Creating Multiplots on Same Canvas.

**Syntax: plt.subplot(nrows, ncols, plot_number)**

```python
# plt.subplot(nrows, ncols, plot_number)

plt.subplot(1, 2, 1)
plt.plot(x, y, 'r*--')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')

plt.show()
```

- <u>NOTE:

    - nrows, ncols
    - How rows and columns translates into the actual figure layout.
    - Example:

                  plt.subplot(1, 2, 1)
                  plt.subplot(1, 2, 2)
                                      |   Plot 1   |   Plot 2   |
      That's one row two plots.
    - Bigger grid Example: plt.subplot(2, 2, i)

                              plt.subplot(2, 2, 1)  # Top-left
                              plt.subplot(2, 2, 2)  # Top-right
                              plt.subplot(2, 2, 3)  # Bottom-left
                              plt.subplot(2, 2, 4)  # Bottom-right

                              +-----------+---------+
                              |  Plot 1  |  Plot 2  |
                              +-----------+---------+
                              |  Plot 3  |  Plot 4  |
                              +-----------+---------+

# <u>2) Object Oriented Method.

- Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object.
- This approach is nicer when dealing with a canvas that has multiple plots on it.

**To begin we create a figure instance. Then we can add axes to that figure:**

```python
# Step 1: Create Figure instance (empty canvas).
fig = plt.figure()
plt.show()
```

```python
# Step 1: Create Figure instance (empty canvas).
fig = plt.figure()

# Step 2: Add set of axes to figure.
axes = fig.add_axes([0.1, 0.1, 1, 1]) # left, bottom, width, height (range 0 to 1)
plt.show()
```

- <u>NOTE:

    - axes = fig.add_axes([0, 0, 0.8, 0.8])  # left, bottom, width, height (range 0 to 1)
    - The list [0.1, 0.1, 0.8, 0.8] is how fig.add_axes() positions the axes inside the figure/empty canvas. Basically the percent of that blank canvas you want to go ahead and take.
    - The first two numbers (left, bottom) will make more sense when we set another figure(Insert axes) inside the bigger figure.
    - left (0.1) → distance from the left edge of the bigger figure(10% of figure width).
    - bottom (0.1) → distance from the bottom edge of the bigger figure(10% of figure height).
    - width (0.8) → how wide the axes are (80% of the figure width).
    - height (0.8) → how tall the axes are (80% of the figure height).

```python
# Step 1: Create Figure instance (empty canvas).
fig = plt.figure()

# Step 2: Add set of axes to figure.
axes = fig.add_axes([0.1, 0.1, 0.2, 0.3]) # left, bottom, width, height (range 0 to 1)
plt.show()
```

```python
# Step 1: Create Figure instance (empty canvas).
fig = plt.figure()

# Step 2: Add set of axes to figure.
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Step 3: Plot on that set of axes.
axes.plot(x, y, 'b*-')
axes.set_xlabel('Set X Label') # Notice the use of set_xlabel instead of just xlabel.
axes.set_ylabel('Set Y Label') # Notice the use of set_ylabel instead of just ylabel.
axes.set_title('Set Title') # Notice the use of set_title instead of just title.

plt.show()
```

### Two sets of figures on one canvas.

And this will start to make sense as far as what this list: [0.1, 0.1, 0.8, 0.8] is referencing.

```python
# Step 1: Create Figure instance (empty canvas).
fig = plt.figure()

# Step 2: Add set of axes to figure.
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
# Add another axes.
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Insert axis

plt.show()
```

- <u>NOTE:

    - axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Insert axis
    - left value = axis starts at 20% out from the left of the bigger figure.
    - bottom value = axis starts at 50% up from the bottom of the bigger figure.
    - width value = takes up 40% of the canvas size in width.
    - height = takes up 30% of the canvas size in height.
    - All values must be between 0 and 1.

### Example: Try to get the insert axis to the bottom right corner.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.1, 0.4, 0.3])

plt.show()
```

- <u>NOTE:

    - The combined width of the entire canvas is 0.1 + 0.8 = 0.9, since axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    - The combined height of the entire canvas is 0.1 + 0.8 = 0.9, since axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    - In order to shift the insert axes to bottom right.
    - We will keep width and height value of insert axis 0.4 and 0.3 for our example.
    - Left value = 0.9 - 0.4 = 0.5
    - Bottom value = 0.1 since we set axes1 bottom value to 0.1.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# 0.5 off
axes2 = fig.add_axes([0.45, 0.15, 0.4, 0.3])

plt.show()
```

### Example: Plot x v/s y for bigger figure and y v/s x for the Insert figure.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Larger Plot axes1
axes1.plot(x, y, 'b*-')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Larger Plot')

# Insert Plot axes2
axes2.plot(y, x, 'g*--')
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('Insert Plot')

plt.show()
```
