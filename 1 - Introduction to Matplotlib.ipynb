- Matplotlib is the most popular plotting library for Python.

- It gives us complete control over almost every aspect of a figure or plot.

- It is an excellent 2D and 3D graphics library for generating scientific figures.

- Matplotlib allows us to create reproducible figures programmatically.

- It's designed to have a very similar feel to Matlab which is another programming language and has its own graphical plotting capabilities.

- Matplotlib works very well with Pandas and NumPy arrays.

- Seaborn is built on top of Matplotlib.

---

# <u>Installation.

We'll need to install matplotlib first with either:

    conda install matplotlib
    or pip install matplotlib

Installed!
