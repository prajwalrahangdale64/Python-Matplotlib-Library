# <u>subplots()

**The plt.subplots() object will act as a more automatic axis manager.**

**Used for creating subplots aranged by rows and columns.**

Notice subplots instead of just subplot.

Basic use cases:

```python
# Step 1: Create a subplots instance/object.
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes.
(fig, axes) = plt.subplots()

plt.show()
```

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes.
(fig, axes) = plt.subplots()


#Step 2: Now use the axes object to add stuff to plot.
axes.plot(x, y, 'r')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')

plt.show()
```

- NOTE:

    - We used to create a figure instance/object by calling plt.figure() then add axes to that instance using fig.add_axes() method.
    - plt.subplots() object acts as a more automatic axis manager.

### We can also specify the number of rows and columns when creating the subplots() object:

**Syntax: (fig, axes) = plt.subplots(nrows, ncols)**

```python
(fig, axes) = plt.subplots(nrows = 1, ncols = 2)

plt.show()
```

- <u>NOTE:

    - The plt.subplots() object acts as a more automatic axis manager essentially calling these (fig, axes) automatically for us based off of these arguments of rows and columns.

### Example: Create a subplot of 3 rows and 3 columns.

```python
(fig, axes) = plt.subplots(nrows = 3, ncols = 3)

plt.show()
```

- <u>NOTE:

    - Notice how the stuff is overlapping.
    - In order to fix this issue we can use a simple one line code: **plt.tight_layout()** at the end of the same cell.

```python
(fig, axes) = plt.subplots(nrows = 3, ncols = 3)

plt.tight_layout()
```

### Example: Revert back to the subplot with 1 row and 2 columns.

```python
(fig, axes) = plt.subplots(nrows = 1, ncols = 2)

plt.show()
```

- <u>NOTE:

    - (fig, axes) is in form of a tuple.

**If we were to tuple unpack axes:**

```python
axes
```

- <u>NOTE:

    - axes is actually just an array of matplotlib axes.

**And we can iterate through axes array to create plots:**

```python
(fig, axes) = plt.subplots(nrows = 1, ncols = 2)

for current_axis in axes:
    current_axis.plot(x, y)

plt.show() # Or just fig if we don't want a print style figure.
```

**We can actually also index axes array to create specific plots:**

```python
(fig, axes) = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y, 'r*-')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')
axes[0].set_title('First Plot')

axes[1].plot(y, x, 'b*--')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')
axes[1].set_title('Second Plot')

plt.show() # OR plt.tight_layout() to avoid overlapping.
```

# <u>Figure size, aspect ratio and DPI.

Matplotlib allows the aspect ratio, DPI and figure size to be specified as arguments when the Figure object is created.

### figsize and dpi keyword arguments.

* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

### Example: Create a x v/s y plot. Specify figsize = (3, 2) and dpi = 100.

```python
# Without specifying figsize = (3, 2) and dpi = 100.

fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y, 'r*-')

plt.show()
```

```python
# With figsize = (8, 2) and dpi = 100 defined.

fig = plt.figure(figsize = (8, 2),  dpi = 100)

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y, 'r*-')

plt.show()
```

```python
# With figsize = (8, 2) and dpi = 100 defined.

fig = plt.figure(figsize = (8, 2),  dpi = 100)

# Altering axes width value to 0.5
axes = fig.add_axes([0, 0, 0.5, 1])

axes.plot(x, y, 'r*-')

plt.show()
```

**The same arguments can also be passed to layout managers, such as the subplots function:**

```python
(fig, axes) = plt.subplots(nrows = 2, ncols = 1,  figsize = (8, 4)) # Not specifying dpi value, going w the default value.

axes[0].plot(x, y, 'b.-')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')
axes[0].set_title('Plot 1')

axes[1].plot(y, x, 'g*--')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')
axes[1].set_title('Plot 2')

plt.tight_layout()
```

```python
# Let's save this previous figure object.

(fig, axes) = plt.subplots(nrows = 2, ncols = 1,  figsize = (8, 4)) # Not specifying dpi value, going w the default value.

axes[0].plot(x, y, 'b*-')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')
axes[0].set_title('Plot 1')

axes[1].plot(y, x, 'g*--')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')
axes[1].set_title('Plot 2')

plt.tight_layout()
```

# <u>Legends, labels and titles.

Now that we have covered the basics of how to create a figure canvas and add axes instances to the canvas, let's look at how decorate a figure with titles, axis labels, and legends.

**We already know how to add titles using `set_title` and Axis labels using `set_xlabel` and `set_ylabel` methods in the axes instance of a figure object.**

### Example: Create a figure object, plot x v/s y and set title, axis labels.

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.65, 0.65])

axes.plot(x, y, 'b')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')

plt.show()
```

### Example: Previous example. Since these line are quite diffrent, plot x v/s xsq and x v/s xcube.

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.65, 0.65])

axes.plot(x, x**2, 'b')
axes.plot(x, x**3, 'g')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')

plt.show()
```

### Legends.

So now we have two lines in the previous example.

And if we were just looking at this figure by itself without being able to reference the code, we'll probably get confused as to what each plot represents.

We would want to add in a legend in order to differentiate the plots.

In order to do that we'll call the axes.legend() method at the bottom of the code.

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.65, 0.65])

axes.plot(x, x**2, 'b')
axes.plot(x, x**3, 'g')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')

# First step.
axes.legend()

plt.show()
```

- <u>NOTE:

    - Notice we get a warning basically stating that no labeled objects were found.
    - In order to fix the warning we need to put in another argument, label in axes.plot() method. 

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.65, 0.65])

axes.plot(x, x**2, 'b', label = 'X Squared') # Second step: label = 'X Squared'
axes.plot(x, x**3, 'g', label = 'X Cubed') # Second step: label = 'X Cubed'
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')

# First step.
axes.legend()

plt.show()
```

- <u>NOTE:

    - Notice that we get the legend now.

    - So all this does is when we call axes.legend(), it looks for each of the plot calls and then checks to see if there's a label.

    - And then the label which is just a string gets returned and gets matched up to the correct color of our line.

**Now sometimes the legend may overlap with some of the actual plot!**

The **legend** function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be drawn. The allowed values of **loc** are numerical codes for the various places the legend can be drawn. See the [documentation page](http://matplotlib.org/users/legend_guide.html#legend-location) for details. Some of the most common **loc** values are:

Lots of options....

- axes.legend(loc=1) # upper right corner
- axes.legend(loc=2) # upper left corner
- axes.legend(loc=3) # lower left corner
- axes.legend(loc=4) # lower right corner
- axes.legend(loc=10) # centre

.. many more options are available

Most common to choose:
- ax.legend(loc=0) # let matplotlib decide the optimal location

Now if none of these options are working for us such as 0 through 10 what we can do is specify a tuple giving the x and y of the lower left corner of the legend. 
- Example: (0.1, 0.1) ---> (width, height)

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.65, 0.65])

axes.plot(x, x**2, 'b', label = 'X Squared') # Second step: label = 'X Squared'
axes.plot(x, x**3, 'g', label = 'X Cubed') # Second step: label = 'X Cubed'
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')

# First step.
axes.legend(loc = 10) # loc = 0, Always use the optimal location option.

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.65, 0.65])

axes.plot(x, x**2, 'b', label = 'X Squared') # Second step: label = 'X Squared'
axes.plot(x, x**3, 'g', label = 'X Cubed') # Second step: label = 'X Cubed'
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')

# First step.
axes.legend(loc = (1.1, 0.8)) # loc = 0, Always use the optimal location option.

plt.show()
```
