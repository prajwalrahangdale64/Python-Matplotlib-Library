# <u>Colors

### Colors with MatLab like syntax.

There is the basic MATLAB like syntax (which we should avoid using for more clairty sake).

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
# MATLAB style line color and style 

(fig, axes) = plt.subplots()

axes.plot(x, x**2, 'b.-') # blue line with dots.
axes.plot(x, x**3, 'g*--') # green dashed line with stars.
```

### Colors with the color= parameter.

We can also define colors by their names(common color strings like: 'blue', 'green' etc) or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, y, color = 'purple') # Defining color= parameter with common color strings.

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, y, color = '#FF8C00') # Defining color= parameter with RGB hex codes.

plt.show()
```

- <u>NOTE:

    - So we can google search for RGB hex code maker or creator and we should have plenty of tools available freely online in case we want a custom color.

```python
(fig, axes) = plt.subplots()

axes.plot(x, x+1, color = 'blue', alpha = 0.5) # using alpha= parameter, half-transparant, range: 0 to 1 
axes.plot(x, x+2, color = '#8B008B')        # RGB hex code
axes.plot(x, x+3, color = '#FF8C00')        # RGB hex code 

plt.show()
```

# <u>Line width and style.

### **1) linewidth**

***To change the line width, we can use the `linewidth` or `lw` keyword argument.***

***Default value for lw = 1***

```python
fig = plt.figure()

axes = fig.add_axes([0,0,0.65,0.65])

axes.plot(x, y, color = 'green', linewidth = 2)

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,0.65,0.65])

axes.plot(x, y, color = 'green', lw = 20) # We can just use the keyword argument lw

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,0.65,0.65])

axes.plot(x, y, color = 'green', lw = 0.5) # If we want a thin line.

plt.show()
```

***Using color, lw and alpha keyword arguments together:***

```python
(fig, axes) = plt.subplots()

axes.plot(x, x+1, color = 'orange', lw = 3, alpha = 0.5) # alpha controls opacity.

plt.show()
```

### **2) linestyle**

***The line style can be selected using the `linestyle` or `ls` keyword arguments:***

```python
(fig, axes) = plt.subplots()


# possible linestyle options '-', '--', '-.', ':'
axes.plot(x, x+1, color = 'red', lw = 1, linestyle = '-')
axes.plot(x, x+2, color = 'green', lw = 2, ls = '--')
axes.plot(x, x+3, color = 'blue', lw = 3, ls = '-.')
axes.plot(x, x+4, color = 'black', lw = 4, ls = ':')
          
plt.show()
```

***If we want a step plot:***

The where parameter decides where the jump (vertical part of the step) happens relative to each x-value(Use 'mid' for most cases):

```python
(fig, axes) = plt.subplots()

# 'pre' = The step changes value before the given x-position.
axes.step(x, x+1, color = 'red', lw = 1, where = 'pre')
# 'mid' = The step changes value after the given x-position.
axes.step(x, x+2, color = 'blue', lw = 2, where = 'mid')
# 'post' = The step changes value halfway between two x-values.
axes.step(x, x+3, color = 'green', lw = 3, where = 'post')

plt.show()
```

# <u>Markers.

**Marks coordinates on lines.**

**Possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4'**

```python
(fig, axes) = plt.subplots(figsize = (12, 6))

axes.plot(x, x+1, color = 'red', lw = 0.5, ls = '-', marker = '+')
axes.plot(x, x+2, color = 'blue', lw = 0.5, ls = '-', marker = 'o') # big dot
axes.plot(x, x+3, color = 'green', lw = 0.5, ls = '-', marker = '*')
axes.plot(x, x+4, color = 'purple', lw = 0.5, ls = '-', marker = 's') # square
axes.plot(x, x+5, color = 'orange', lw = 0.5, ls = '-', marker = ',')
axes.plot(x, x+6, color = 'pink', lw = 0.5, ls = '-', marker = '.') # small dot
axes.plot(x, x+7, color = 'black', lw = 0.5, ls = '-', marker = '1')


plt.show()
```

### Marker size and color.

**Keyword arguments:**

1) markersize = affects the size of the marker.
2) markerfacecolor = affects the color of the marker.
3) markeredgewidth = affects the edge width of the marker.
4) markeredgecolor = affects the color of the marker edge.

```python
(fig, axes) = plt.subplots(figsize = (12, 6))

axes.plot(x, x+1, color = 'red', lw = 1, ls = '-', marker = '+', markersize = 10)
axes.plot(x, x+2, color = 'blue', lw = 1, ls = '-', marker = 'o', markersize = 10, markerfacecolor = 'yellow', 
          markeredgewidth = 2, markeredgecolor = 'green')
axes.plot(x, x+3, color = 'green', lw = 0.5, ls = '-', marker = '*',markersize = 10, markerfacecolor = 'red', 
          markeredgewidth = 1, markeredgecolor = 'black')
axes.plot(x, x+4, color = 'green', lw = 1, ls = '-', marker = 's', markersize = 10, markerfacecolor = 'orange', 
          markeredgewidth = 3, markeredgecolor = 'red')
axes.plot(x, x+5, color = 'yellow', lw = 0.5, ls = '-', marker = ',', markersize = 20, markerfacecolor = 'black')
axes.plot(x, x+6, color = 'orange', lw = 1, ls = '-', marker = '.', markersize = 10)
axes.plot(x, x+7, color = 'pink', lw = 1, ls = '-', marker = '1', markersize = 10)


plt.show()
```

# <u>Control over axis appearance.

In this section we will look at controlling axis sizing properties in a matplotlib figure.

# <u>Plot range.

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
x
```

```python
x**2
```

```python
x**3
```

```python
(fig, axes) = plt.subplots(nrows = 1, ncols = 3, figsize = (12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title('default axes ranges')
axes[0].set_xlabel('x')
axes[0].set_ylabel('x**2 or x**3')

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title('tight axes')

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60]) # [lowerbound, upperbound]
axes[2].set_xlim([2, 5]) # [lowerbound, upperbound]
axes[2].set_title('custom axes range')

plt.show()
```

```python
y
```

***Another example.***

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.65, 0.65])

axes.plot(x, y, color = 'purple', lw = 2, ls = '--')
axes.set_title('default axes ranges')

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.65, 0.65])

axes.plot(x, y, color = 'purple', lw = 2, ls = '--')
axes.set_ylim([0, 1])
axes.set_xlim([0, 2])
axes.set_title('custom axes ranges')

plt.show()
```

- <u>NOTE:

    - And notice as we zoom in the plot, it looks more and more like a bunch of straight lines connecting data points which makes sense because we don't have as much detail since we only have 11 data points.
    - We can fix this by getting more data points or by zooming out.
    - As we zoom out, we will see more of a curved line.

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.65, 0.65])

axes.plot(x, y, color = 'purple', lw = 2, ls = '--')
axes.axis('tight')
axes.set_title('tight axes')

plt.show()
```

# <u>Special Plot Types.

There are many specialized plots we can create, such as barplots, histograms, scatter plots, and much more. Most of these type of plots we will actually create using seaborn, a statistical plotting library for Python. But here are a few examples of these type of plots:

### 1) Scatter plots.

```python
plt.scatter(x, y)

plt.show()
```

```python
fig = plt.figure(figsize = (8, 8))

axes1 = fig.add_axes([0, 0, 1, 1]) 
axes2 = fig.add_axes([0.1, 0.45, 0.5, 0.5]) 

axes1.scatter(x, y, color = 'blue')
axes1.set_title('x v/s y')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')

axes2.scatter(y, x, color = 'red', alpha = 0.5, marker = '+', label = 'y v/s x')
axes2.scatter(x, x**2, color = 'green', alpha = 0.5, marker = 's', label = 'x squared') # Can take color, alpha, and marker argumets also.
axes2.set_title('Insert Plot')
axes2.legend()

plt.show()
```

**Another example:**

```python
(fig, axes) = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 6))

axes[0].scatter(x, y, color = 'blue')
axes[0].set_title('x v/s y')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')

axes[1].scatter(y, x, color = 'red', alpha = 0.5, marker = '+', label = 'y v/s x')
axes[1].scatter(x, x**2, color = 'green', alpha = 0.5, marker = 's', label = 'x squared') # Can take color, alpha, and marker argumets also.
axes[1].set_ylim([0, 20])
axes[1].set_title('2nd Plot')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')
axes[1].legend()

plt.show()
```

### 2) Histograms.

***Something i missed in NumPy:***

```python
# np.arange(start, stop)
# This was covered in NumPy.

np.arange(1, 10)
```

```python
# np.random.randint(low(inclusive), high(exclusive), size)
# This was covered in NumPy.
# For a 1-D array.
# For a 2-D array: size = (rows, columns)

np.random.randint(1, 1000, 100)
```

- <u>NOTE:

    - np.random.randint(low(inclusive), high(exclusive), size)
    - It samples with replacement.
    - Duplicates are allowed.

***In order to get random numbers without replacement.***

***We can use np.random.choice() function.***

**Syntax: np.random.choice(np.arange(start, stop), size, replace = True)**

        where;
                size = None  number of samples.
                replace = False   without replacement (True by default).

```python
# replace = False ---> without replacement.
# replace = True ---> with replacement.

np.random.choice(np.arange(1, 1000), 100, replace = False)
```

***Back to histograms:***

***How are Histograms plotted in general?***





***Using Matplotlib:***

**Syntax: plt.hist(data, bins)**

    where;
        bins = 10 by default (10 equal-width bins).

```python
data = np.random.choice(np.arange(1, 1000), 100, replace = False)
```

- <u>NOTE:

    - For default bin size(10).
    - np.arange(1, 1000)
    - Our data is between 1 and 999.
    - Range = max - min = 999  1 = 998.
    - Divide this range into 10 equal parts  each bin width  99.8  100.
    - So bins will look like:
    - 1101, 101201, 201301, ... , 901999

```python
# With default bin size.

plt.hist(data)

plt.show()
```

Custom bins:

You can control bins in two ways:

By number of bins and By exact bin edges.

```python
# By number of bins.
# With custom equal width bin size.

plt.hist(data, bins = 10)

plt.show()
```

```python
# By number of bins.
# With custom equal width bin size.

plt.hist(data, bins = 20)

plt.show()
```

- <u>NOTE:

    - bins = 20
    - np.arange(1, 1000)
    - Our data is between 1 and 999.
    - Range = max - min = 999  1 = 998.
    - Divide this range into 20 equal parts  each bin width  49.9  50.
    - So bins will look like:
    - 151, 51101, 101151, ... , 951999

```python
# By exact bin edges:
# Bins(Intervals): 0-100, 100-200, 200-500, 500-999

plt.hist(data, bins = [0, 100, 200, 500, 1000])

plt.show()
```

***Another example:***

```python
(fig, axes) = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 6))

axes[0].hist(data, bins = 10, color = 'red')
axes[0].set_title('Bin size 10')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')

axes[1].hist(data, bins = 20, color = 'blue', alpha = 0.5, label = 'Bin size 20')
axes[1].hist(data, bins = 30, color = 'green', alpha = 1, label = 'Bin size 30') # Can take color and alpha argumets.
axes[1].set_title('2nd Plot')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')
axes[1].legend()

plt.show()
```

***Another example:***

```python
(fig, axes) = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 6))

axes[0].hist(data, bins = 10, color = 'red')
axes[0].set_title('Bin size 10')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')

axes[1].hist(data, bins = 20, color = 'blue', alpha = 0.5, label = 'Bin size 20')
axes[1].hist(data, bins = 30, color = 'green', alpha = 1, label = 'Bin size 30') # Can take color, alpha, and marker argumets also.
axes[1].set_xlim([0, 600])
axes[1].set_title('2nd Plot')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')
axes[1].legend(loc = (1.1, 1))

plt.show()
```

### 3) Boxplot.

    Draws a box and whisker plot.

***Creating Data:***

```python
# Generate 100 values from normal distribution.
# Using a custom mean () and standard deviation ().
# In a 1-D array.
# loc = 0  mean ()
# scale = 1  standard deviation ()
# size = 100  number of samples
# size = 100 ---> Shape(100, ) --> 100 elements in one dimension.


np.random.normal(loc = 0, scale = 1, size = 100)
```

```python
# List Comprehensions.

mylist = []

for item in range(1, 4):
    mylist.append(item)

mylist
```

```python
# List Comprehensions (Alternate method).

mylist1 = [item for item in range(1, 4)]
mylist1
```

```python
# List Comprehensions.

data1 = []

for std in range(1, 4):
    data1.append(np.random.normal(loc = 0, scale = std, size = 100))

data1
```

```python
# List Comprehensions (Alternate method).

data1 = [np.random.normal(loc = 0, scale = std, size = 100) for std in range(1, 4)]
data1
```

***Boxplot in general:***

A boxplot shows:

    - Center (median),

    - Spread (IQR, whiskers),

    - Skewness (if median is off-center inside the box),

    - Outliers (points beyond whiskers).



1) Median (Q2):

    The line inside the box is the median (50th percentile).

2) Quartiles (Q1 and Q3):

    Q1 (25th percentile): the lower edge of the box.

    Q3 (75th percentile): the upper edge of the box.

    So the box spans from Q1 to Q3.





4) Interquartile Range (IQR):

    IQR=Q3Q1

    Represents the middle 50% of the data.

5) Whiskers: Extend outward from Q1 and Q3 (lowest and highest value).

    1.5  

6) Outliers:

    Any data points beyond the whiskers are plotted individually (often as dots or small circles).

```python
# Rectangular box plot.

plt.boxplot(data1, vert = True, patch_artist = True)

plt.show()
```

```python
(fig, axes) = plt.subplots(figsize = (12, 6))

axes.boxplot(data1, vert = True, patch_artist = True, label = 'Test')
axes.set_title('Boxplot')
axes.set_xlabel('Std. deviation')
axes.set_ylabel('Random Normal Values (-, +)')

axes.legend()

plt.show()
```

```python
(fig, axes) = plt.subplots(figsize = (12, 6))

axes.boxplot(data1, vert = True, patch_artist = True, label = 'Test')
axes.set_title('Boxplot')
axes.set_xlabel('Std. deviation')
axes.set_ylabel('Random Normal Values (-, +)')
axes.set_ylim([-8, 5])

axes.legend()

plt.show()
```
